### Part 1: Pre-processing and EDA

#### Problem 1: Dataset Importing and Cleaning

```python
orders = pd.read_csv('Orders.csv')
returns = pd.read_csv('Returns.csv')

orders
```

```python
orders['Profit'] = orders.Profit.replace('[$,]', '', regex=True).astype(float)
orders['Sales'] = orders.Profit.replace('[$,]', '', regex=True).astype(float)
```

#### Problem 2: Inventory Management

```python
orders['Month'] = pd.to_datetime(orders['Order.Date']).dt.month
```

```python
orders
```

```python
seasons = {1: 'Winter', 2: 'Winter', 3: 'Spring', 4: 'Spring', 5: 'Spring', 6: 'Summer', 7: 'Summer', 8: 'Summer', 9: 'Fall', 10: 'Fall', 11: 'Fall', 12: 'Winter'}
```

```python
orders['Season']= orders['Month'].map(seasons)
```

```python
orders['Season']
```

```python
df = orders.groupby(['Order.Date'])['Quantity'].mean()

df.plot.bar()
```

Graph above does support some indication of a seasonal trend. 

```python
orders['Category'].unique()
```

```python
df3 = orders.groupby(['Category', 'Order.Date'])['Quantity'].mean()

df3.plot.bar()
```

There does seem to be a seasonal trend for different categories.
